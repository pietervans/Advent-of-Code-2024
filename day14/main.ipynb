```python
import re
import math
import numpy as np
from tqdm import tqdm

with open('input.txt') as f:
    lines = f.readlines()

positions = []
velocities = []
for l in lines:
    m = re.search('p=(\d+),(\d+) v=(-?\d+),(-?\d+)', l)
    positions.append( (int(m.group(1)), int(m.group(2))) )
    velocities.append( (int(m.group(3)), int(m.group(4))) )
```

```python
I = 101
J = 103

def plus(pos1, pos2):
    return ((pos1[0]+pos2[0]) % I, (pos1[1]+pos2[1]) % J)
def mult(pos, n):
    return (pos[0]*n, pos[1]*n)

iteration_scores = []

def print_picture(positions):
    grid = [[' ' for j in range(J)] for i in range(I)]
    for pos in positions:
        grid[pos[0]][pos[1]] = '*'
    for row in grid:
        print(''.join(row))

for k in tqdm(range(I*J)):
    # print(k)
    new_positions = []
    for pos, vel in zip(positions, velocities):
        new_pos = plus(pos, vel)
        new_positions.append(
            new_pos
        )
    
    positions = new_positions
    
    if k == 7583:
        print_picture(positions)
    
    scores = [0]*4
    for pos in positions:
        i,j = pos
        if i<I//2 and j<J//2:
            scores[0] += 1
        
        if i>I//2 and j<J//2:
            scores[1] += 1
        
        if i<I//2 and j>J//2:
            scores[2] += 1
        
        if i>I//2 and j>J//2:
            scores[3] += 1

    iteration_scores.append(math.prod(scores))

iteration_scores[1119] = float('inf')
min_score = min(iteration_scores)
iteration_scores.index(min_score)
```
