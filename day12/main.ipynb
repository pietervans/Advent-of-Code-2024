```python
with open('input.txt') as f:
    lines = f.readlines()

grid = [s.strip() for s in lines]
```

```python
I = len(grid)
J = len(grid[0])
    
def is_inside_grid(position):
    return position[0] in range(I) and position[1] in range(J)

def get_neighbours(position):
    return set([
        (position[0]+1, position[1]),
        (position[0]-1, position[1]),
        (position[0],   position[1]+1),
        (position[0],   position[1]-1)
    ])

def get_region(position, current_region=None):
    if not current_region:
        current_region = set([position])
    regions_temp = []
    for n in get_neighbours(position):
        if is_inside_grid(n) and \
            grid[n[0]][n[1]] == grid[position[0]][position[1]] and \
            n not in current_region:
            current_region.add(n)
            regions_temp.append(get_region(n, current_region))
    res = current_region
    for r in regions_temp:
        res = res | r
    return res

def get_area(region):
    return len(region)

def get_perimeter(region):
    res = 0
    for p in region:
        res += 4 - len(region & get_neighbours(p))
    return res

def get_perimeter_sections(region):
    '''Get list of perimeter sections.
    Every element is (position_in, position_out)'''
    res = []
    for p in region:
        for n in get_neighbours(p):
            if n not in region:
                res.append((p, n))
    return res

def is_on_same_fence_side(sec1, sec2):
    in1, out1 = sec1
    in2, out2 = sec2
    a1, b1 = in1
    c1, d1 = out1
    a2, b2 = in2
    c2, d2 = out2
    if a1==a2 and c1==c2:
        if abs(b1-b2)==1 and abs(d1-d2)==1:
            return True
    if b1==b2 and d1==d2:
        if abs(a1-a2)==1 and abs(c1-c2)==1:
            return True
    return False

def get_nb_sides(region):
    sections = get_perimeter_sections(region)
    sections_accounted_for = []
    section_groups = []
    for sec1 in sections:
        if sec1 in sections_accounted_for:
            continue
        section_group = [sec1]
        change = True
        while change:
            change = False
            for section_from_group in section_group:
                
                for sec2 in sections:
                    if is_on_same_fence_side(section_from_group, sec2):
                        if sec2 not in sections_accounted_for:
                            change = True
                            section_group.append(sec2)
                            sections_accounted_for.append(sec2)                        
        section_groups.append(section_group)
    # print(section_groups)
    return len(section_groups)

res = 0
positions_accounted_for = set()
for i in range(I):
    for j in range(J):
        pos = (i,j)
        if pos not in positions_accounted_for:
            region = get_region(pos)
            positions_accounted_for = positions_accounted_for | region
            res += get_area(region)*get_nb_sides(region)
            print(grid[i][j], get_area(region), get_nb_sides(region))

# r = get_region((0,0))
# get_nb_sides(r)
res
```
