```python
with open('input.txt') as f:
    lines = f.readlines()

test_values = dict()
for l in lines:
    test_value, nums = l.split(': ')
    test_values[int(test_value)] = [int(s) for s in nums.strip().split(' ')]
```

```python
def calculate(nums, binary_operators):
    calc = nums[0]
    for n, op in zip(nums[1:], binary_operators):
        if op == '0':
            calc += n
        else:
            calc *= n
    return calc

res = 0
for test_value, nums in test_values.items():
    for i in range(2**(len(nums)-1)):
        operators = bin(i)[2:].rjust(len(nums)-1, '0')
        c = calculate(nums, operators)
        # print(test_value, nums, operators, c)
        if c == test_value:
            res += test_value
            break
res
```

```python
# Part 2
import numpy as np

def calculate(nums, binary_operators):
    calc = nums[0]
    for n, op in zip(nums[1:], binary_operators):
        if op == '0':
            calc += n
        elif op == '1':
            calc *= n
        else:
            calc = int(str(calc)+str(n))
    return calc



res = 0
k = 1
for test_value, nums in test_values.items():
    print(f'{k}/{len(test_values)}'); k+= 1
    for i in range(3**(len(nums)-1)):
        operators = np.base_repr(i, base=3).rjust(len(nums)-1, '0')
        c = calculate(nums, operators)
        # print(test_value, nums, operators, c)
        if c == test_value:
            res += test_value
            break
res
```
