```python
import re
import numpy as np


with open('input.txt') as f:
    lines = f.readlines()

jumps_A = []
jumps_B = []
prizes = []

for line in lines:
    match_A = re.search('Button A: X\+(\d+), Y\+(\d+)', line)
    match_B = re.search('Button B: X\+(\d+), Y\+(\d+)', line)
    match_prize = re.search('Prize: X=(\d+), Y=(\d+)', line)
    if match_A:
        jumps_A.append((int(match_A.group(1)), int(match_A.group(2))))
    if match_B:
        jumps_B.append((int(match_B.group(1)), int(match_B.group(2))))
    if match_prize:
        # prizes.append((int(match_prize.group(1)), int(match_prize.group(2))))
        prizes.append((10000000000000+int(match_prize.group(1)), 10000000000000+int(match_prize.group(2))))
```

```python
def mult(pos, n):
    return (pos[0]*n, pos[1]*n)

def plus(pos1, pos2):
    return (pos1[0]+pos2[0], pos1[1]+pos2[1])

def is_close_to_integer(f):
    return abs(f-round(f)) < 0.001

total = 0

for i in range(len(prizes)):
    print(i)
    
    jump_A = jumps_A[i]
    jump_B = jumps_B[i]
    prize = prizes[i]
    
    # res = None
    # for m in range(101):
    #     for n in range(101):
    #         if plus(mult(jump_A, m), mult(jump_B, n)) == prize:
    #             print('Brute ', (m, n))
    #             res = min(res if res else 9999999999, 3*m + n)
    
    # if res:
    #     total += res
     
    
    a = np.matrix([
        [jump_A[0], jump_B[0]],
        [jump_A[1], jump_B[1]]
    ])
    b = np.matrix([
        [prize[0]],
        [prize[1]]
    ])
    x = np.linalg.solve(a, b)
    if is_close_to_integer(x[0,0]) and is_close_to_integer(x[1,0]):
        # print(x[0,0], x[1,0])
        # if x[0,0] >= 0 and x[1,0] >= 0:
        total += 3*round(x[0,0]) + round(x[1,0])
    
    # print('System', (x[0,0], x[1,0]))

total
```
