```python
data_expanded = []
for i, e in enumerate(data):
    if i%2 == 0:
        # file
        data_expanded += [i//2 for k in range(int(e))]
    else:
        # free
        data_expanded += [-1 for k in range(int(e))]
# data_expanded
```

```python
## Part 2

file_id_max = data_expanded[-1]
file_id_move = data_expanded[-1]+1
while file_id_move > -1:
    
    print(f'{file_id_move}/{file_id_max}')
    
    # Find next file to try to move
    file_id_move -= 1
    id_index_end = data_expanded[::-1].index(file_id_move)
    len_file = 1
    for id in data_expanded[::-1][id_index_end+1:]:
        if id != file_id_move:
            break
        else:
            len_file += 1
        
    
    # Look for next chunk of free space that fits file
    free_space_index_stop = -1
    while True:
        try:
            free_space_index_start = data_expanded.index(-1, free_space_index_stop+1)
        except:
            break
        num_free_spaces = 1
        free_space_index_stop = free_space_index_start
        for x in data_expanded[free_space_index_start+1:]:
            if x != -1:
                break
            num_free_spaces += 1
            free_space_index_stop += 1
        if num_free_spaces >= len_file:
            break
    
    
    if num_free_spaces < len_file:
        continue
    # print(f'free_space_index_start={free_space_index_start}')
    # print(f'num_free_spaces={num_free_spaces}')
    
    
        
    # print(f'file_id_move={file_id_move}')
    # print(f'len_file={len_file}')
        
        
        
    if free_space_index_start > len(data_expanded)-1-id_index_end-len_file+1:
        # Would not move file left
        continue
    
    # Move file
    for i in range(len_file):
        data_expanded[free_space_index_start] = file_id_move
        free_space_index_start += 1
        data_expanded[-1-id_index_end] = -1
        id_index_end += 1
    

# print('   '.join([str(n) for n in data_expanded]))
print(data_expanded)
```
